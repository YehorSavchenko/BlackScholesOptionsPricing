```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def black_scholes_call(S, X, T, r, sigma):
    d1 = (np.log(S / X) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call = S * norm.cdf(d1) - X * np.exp(-r * T) * norm.cdf(d2)
    return call
```

```python
S = 100  # Current stock price
X = 100  # Strike price
T = 1    # Time to expiration (in years)
r = 0.05 # Risk-free interest rate
sigma = 0.2 # Volatility

call_price = black_scholes_call(S, X, T, r, sigma)
print(f"The price of the call option is: {call_price:.2f}")
```

```python
volatilities = np.linspace(0.1, 0.5, 100)
call_prices = [black_scholes_call(S, X, T, r, v) for v in volatilities]

plt.plot(volatilities, call_prices)
plt.xlabel('Volatility')
plt.ylabel('Call Option Price')
plt.title('Call Option Price vs Volatility')
plt.show()
```
